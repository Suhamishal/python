```python
import random

##Chromosome length = grid size
grid_size=6
lower_bound=1
upper_bound=6
population_size = 20
fitnessvalues=[]
generations_ff=[]

##Randomly generated set of 5 genes (x,y coordinates) for 20 chromosomes
population=([(4, 4), (3, 1), (1, 4), (4, 2), (5, 2)], [(4, 4), (6, 2), (5, 5), (2, 1), (6, 5)], [(3, 4), (5, 4), (2, 3), (1, 6), (6, 3)], [(6, 2), (1, 2), (5, 5), (2, 4), (3, 2)], [(4, 4), (4, 2), (4, 5), (3, 6), (1, 6)], [(2, 1), (5, 1), (4, 2), (5, 3), (4, 1)], [(2, 1), (4, 3), (4, 5), (5, 3), (1, 6)], [(6, 1), (1, 1), (2, 5), (1, 3), (3, 5)], [(4, 3), (2, 2), (5, 3), (4, 1), (3, 5)], [(4, 3), (5, 3), (6, 3), (1, 3), (5, 2)], [(4, 4), (1, 5), (6, 4), (3, 2), (5, 2)], [(5, 4), (5, 1), (1, 6), (3, 2), (3, 5)], [(1, 2), (6, 4), (4, 2), (5, 6), (1, 6)], [(5, 5), (2, 1), (4, 3), (4, 1), (3, 5)], [(1, 1), (5, 4), (5, 3), (6, 3), (1, 3)], [(2, 4), (5, 4), (3, 6), (2, 5), (4, 1)], [(2, 4), (4, 3), (5, 1), (3, 3), (6, 3)], [(6, 2), (4, 6), (2, 3), (6, 6), (4, 1)], [(2, 1), (1, 1), (2, 3), (4, 5), (3, 6)], [(3, 4), (4, 3), (5, 4), (5, 1), (3, 2)])


##Function to calculate Fitness
def calculate_fitness(chromosome):

    # Initialize a set to store the covered coordinates
    antenna_coverage = set()

    # Iterate over each antenna in the chromosome
    for antenna in chromosome:
        x, y = antenna

        # Include the block on which the coordinate is located
        antenna_coverage.add((x, y))

        # Including four neighbors (up, down, left, right)
        neighbors = [(x - 1, y), (x + 1, y), (x, y - 1), (x, y + 1)]

        # Ensure that the neighbors' coordinates are within the 6x6 grid
        neighbors = [(i, j) for i, j in neighbors if 1 <= i <= 6 and 1 <= j <= 6]

        # Include neighbors in the coverage set
        antenna_coverage.update(neighbors)

    # Return the count of covered coordinates as the fitness value
    return len(antenna_coverage)

##Initialize an empty list to store fitness values
fitnessvalues = []

##Initialize an empty set to store the covered coordinates
covered_coordinates = set()

##Iterate over the population and calculate fitness values
for i, chromosome in enumerate(population):
    fitness_value = calculate_fitness(chromosome)
    fitnessvalues.append(fitness_value)
    if i == 0:
        covered_coordinates.update(set(chromosome))
    else:
        covered_coordinates.update(set(chromosome).intersection(covered_coordinates))

##Display the initial population with fitness values
print("Initial Population:")
for i in range(population_size):
    print(f"Chromosome {i + 1}: {population[i]} - Fitness Value: {fitnessvalues[i]}")

##Combine population and fitnessvalues
population_with_fitness = list(zip(population, fitnessvalues))

##Sort in descending order based on fitness values
population_with_fitness.sort(key=lambda x: x[1], reverse=True)

##Separate sorted population and fitnessvalues
sorted_population, fitnessvalues = zip(*population_with_fitness)

##Display the sorted population
print("\nSorted Population:")
for i in range(population_size):
    print(f"Chromosome {i + 1}: {sorted_population[i]} - Fitness Value: {fitnessvalues[i]}")


##Take the two chromosomes with the lowest fitness as parents, as we've sorted the chromosomes, so the lowest two will be in the last indexes
parent1_2A=sorted_population[-1]
parent2_2A=sorted_population[-2]

##Crossing over between the parent 1 and 2 for Generation 2A
def crossover(parent1, parent2):
    crossover_point = len(parent1) // 2
    child = parent1[:crossover_point] + parent2[crossover_point:]
    print("The crossover between the chromosome",parent1,"and",parent2,"gives us child",child)
    return child


child1=crossover(parent1_2A,parent2_2A)
child2=crossover(parent2_2A,parent1_2A)

child1_2A_ff=calculate_fitness(child1)
child2_2A_ff=calculate_fitness(child2)
print("Fitness of child 1 of 2A is: ",child1_2A_ff)
print("Fitness of child 2 of 2A is: ",child2_2A_ff)


#Now crossing over between child 1 and the 1st fittest chromosome
child3=crossover(child1,sorted_population[0])
child4=crossover(sorted_population[0],child1)
child3_2A_ff=calculate_fitness(child3)
child4_2A_ff=calculate_fitness(child4)

print("Fitness of child 3 of 2A is: ",child3_2A_ff)
print("Fitness of child 4 of 2A is: ",child4_2A_ff)

#Now crossing over between child 3 and child 1

child5=crossover(child3,child1)
child6=crossover(child1,child3)
child5_2A_ff=calculate_fitness(child5)
child6_2A_ff=calculate_fitness(child6)

print("Fitness of child 5 of 2A is: ",child5_2A_ff)
print("Fitness of child 6 of 2A is: ",child6_2A_ff)

#Now crossing over between chromosome 10 and child 6
child7=crossover(sorted_population[10],child6)
child8=crossover(child6,sorted_population[10])
child7_2A_ff=calculate_fitness(child7)
child8_2A_ff=calculate_fitness(child8)

print("Fitness of child 7 of 2A is: ",child7_2A_ff)
print("Fitness of child 8 of 2A is: ",child8_2A_ff)

#Since fitness function is not changing as such, we will introduce mutation on the chromosome with lower fitness function which is
# on child 7

def change_antenna_coordinates(child, index_to_change, new_x, new_y):

    # Make a copy of the child to avoid modifying the original list
    modified_child = list(child)
    # Change the coordinates of the selected antenna
    modified_child[index_to_change] = (new_x, new_y)
    return modified_child

original_child7 = [(4, 4), (6, 2), (5, 5), (2, 4), (3, 2)]
index_to_change = 3
new_x_coordinate = 2
new_y_coordinate = 5

modified_child = change_antenna_coordinates(original_child7, index_to_change, new_x_coordinate, new_y_coordinate)

# Display the modified child
print("The mutated child is", modified_child,"with fitness function", calculate_fitness(modified_child))
```
